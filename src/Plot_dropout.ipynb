```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(rc={'figure.facecolor': (1,1,1,0)})

files = ['saved_history/dropout0.txt','saved_history/dropout1.txt','saved_history/dropout3.txt','saved_history/dropout5.txt']
plt.figure(figsize=(12,7))
for filename in files:
    train_loss = []
    train_acc = []
    test_loss = []
    test_acc = []
    with open(filename, 'r') as outfile:
        for line in outfile.readlines()[1::2]:
            useful_info = line.split(":")[1:]
            train_loss.append(float(useful_info[0].split()[0]))
            train_acc.append(float(useful_info[1].split()[0]))
            test_loss.append(float(useful_info[2].split()[0]))
            test_acc.append(float(useful_info[3].split()[0]))
    #         print(line)
    plt.plot(test_acc,label = 'dropout = 0.'+filename[-5])
#         plt.plot(test_acc)

# plt.title('RCNN Model Testing Accuracy',fontsize=16)
plt.ylim(0,1)
plt.ylabel('Testing accuracy',fontsize=16)
plt.xlabel('Epoch',fontsize=16)
plt.legend(fontsize=16)
plt.savefig("dropout_test_acc.png",fontsize=16)
plt.show()
```

```python
filename = 'saved_history/dropout5.txt'
```

```python
train_loss = []
train_acc = []
test_loss = []
test_acc = []
with open(filename, 'r') as outfile:
    for line in outfile.readlines()[1::2]:
        useful_info = line.split(":")[1:]
        train_loss.append(float(useful_info[0].split()[0]))
        train_acc.append(float(useful_info[1].split()[0]))
        test_loss.append(float(useful_info[2].split()[0]))
        test_acc.append(float(useful_info[3].split()[0]))
#         print(line)
```

```python
%matplotlib inline  
import matplotlib.pyplot as plt
# summarize history for accuracy
plt.plot(train_acc)
plt.plot(test_acc)
plt.title('RCNN Model Accuracy with Dropout = 0.5')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.savefig("dropout5_acc.png")
plt.show()
# summarize history for loss
plt.figure()
plt.plot(train_loss)
plt.plot(test_loss)
plt.title('RCNN Model Loss with Dropout = 0.5')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
